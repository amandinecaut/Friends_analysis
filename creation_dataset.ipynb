# Create the dataset for The Big Bang Theory

```python
df = pd.read_csv("tbbt.csv", delimiter=',', encoding='utf8') 
df.rename(columns = {'person_scene':'Person', 'dialogue':'Said'}, inplace = True)
df = df[['Person', 'Said']]
```

```python
df
```

```python
# Keep the main character
persons = ['Sheldon', 'Leonard', 'Raj', 'Penny','Howard','Amy','Bernadette']
data = df[df.Person.isin(persons)]
print(data)
```

```python
# Drop empty line
data = data.dropna()
data = data.reset_index(drop=True)

print(data)
```

```python
import csv
import json
import pickle
import pandas as pd
from sentence_transformers import SentenceTransformer,util
import time

# Create the embedding
model = SentenceTransformer('all-MiniLM-L6-v2')
start_time = time.time()


new_column_values = []
for index in range(len(data)):
    print(index)
    print(data.iloc[index][1])   
    array_value = model.encode(data.iloc[index][1])
    new_column_values.append(array_value)
data['Embedding'] = new_column_values

#save the dataset 
#data.to_pickle('C:../data_TBBT_emb.pkl')

end_time = time.time()
execution_time = end_time - start_time

print("Execution time:", execution_time, "seconds")
```
